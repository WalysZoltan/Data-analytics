```python
#Напишите функцию, которая вычленяет из текста все числа и возвражает их списком

#Пробегаясь по всем символам строки, при встрече цифры запоминаю её. Если следующий символ тоже цифра, запоминаю
#далее, иначе понимаю, что поток цифр окончен и делаю запись в лист.

def take_numb(my_string):
    lst = list()
    num_str = ''
    for i in range(len(my_string)):
        if my_string[i] >= '0' and my_string[i] <= '9':
            num_str = num_str + my_string[i]
        else:
            if num_str != '':
                lst.append(int(num_str))
                num_str = ''
            continue
    if num_str != '':
        lst.append(int(num_str))    
    return(lst)

print(take_numb('I was born in 1998'))  #[1998]
print(take_numb('She had got 5 cats and 3 kittens')) #[5, 3]
print(take_numb('Does it really cost 500$? Or maybe 450$?')) #[500, 450]
print(take_numb('lllll'))
print(take_numb('345gghy53yhnj67jum45reb01gf'))
```

```python
#Посчитайте процент выживаемости для мужчин и женщин
overall = df.groupby('Sex').count()['PassengerId']
surv = df.groupby('Sex').sum()['Survived']
print(surv / overall * 100)
```

```python
#Выведите самое популярное мужское и самое популярное женское имя на корабле

#Для мужчин именем считается первое слово после встреченной точки. Для женщин так же, если нет имени в скобках,
#иначе - первое слово в скобке.

def pop_male(x_np):
    male = dict()
    for stri in x_np:
        male[stri.split('.')[1].split()[0].strip()] = male.get(stri.split('.')[1].split()[0].strip(), 0) + 1
        #print(str.split('.')[1].split()[0].strip())
    print("Popular male\n", max(male, key=male.get), max(male.values()), sep = ' ')

def pop_female(x_np):
    female = dict()
    for stri in x_np:
        flag = 0
        for char in stri:
            if char == '(':
                flag = 1
                break;
        if flag == 0:
            female[stri.split('.')[1].split()[0].strip()] = female.get(stri.split('.')[1].split()[0].strip(), 0) + 1
        else:
            female[stri.split('(')[1].split()[0].strip()] = female.get(stri.split('(')[1].split()[0].strip(), 0) + 1
    print("Popular female\n", max(female, key=female.get), max(female.values()), sep = ' ')

pop_male(df[df['Sex'] == 'male']['Name'].values)
pop_female(df[df['Sex'] == 'female']['Name'].values)
```

```python
#Выведите самое популярное мужское и самое популярное женское имя на корабле в каждом классе
for i in range(1, 4):
    print("Pclass", i)
    pop_male(df[(df['Sex'] == 'male') & (df['Pclass'] == i)]['Name'].values)
    pop_female(df[(df['Sex'] == 'female') & (df['Pclass'] == i)]['Name'].values)
```
